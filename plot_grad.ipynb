```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import scipy
# 将PyTorch张量转换为NumPy数组
numpy_array = np.load('grad.npy')

# 使用Matplotlib绘制直方图
figs, axs = plt.subplots(10, 10, figsize=(50, 50))
for i in tqdm(range(100)):
    axs[i//10, i%10].hist(numpy_array[:,i,0], bins=100, density=True, alpha=0.5)
    
    #使用正态分布拟合并画出拟合曲线
    mu, sigma = scipy.stats.norm.fit(numpy_array[:,i,0])
    x = np.linspace(-40, 40, 100)
    y = scipy.stats.norm.pdf(x, mu, sigma)
    axs[i//10, i%10].plot(x, y, 'r--', linewidth=2)
        
    axs[i//10, i%10].set_xlim(-40, 40)
    
    # axs[i//10, i%10].set_title('{.2f},{.2f}'.format(float(mu), float(sigma)))
    
    axs[i//10, i%10].set_title('{:.2f},{:.2f}'.format(float(mu), float(sigma)))
    
    # axs[i//10, i%10].axes.get_xaxis().set_visible(False)
    # axs[i//10, i%10].axes.get_yaxis().set_visible(False)
    # axs[i//10, i%10].set_xlabel('Value')
    # axs[i//10, i%10].set_ylabel('Frequency')
    # axs[i//10, i%10].set_title('Data Distribution of PyTorch Tensor')
# plt.hist(numpy_array[:,0,0], bins=100, density=False, alpha=0.3)
# plt.hist(numpy_array[:,1,0], bins=100, density=False, alpha=0.3)
# plt.xlabel('Value')
# plt.ylabel('Frequency')
# plt.title('Data Distribution of PyTorch Tensor')
plt.savefig('grad.jpg')
```
